```python
def book_scraper(url):
    response = requests.get(url)
    page = TextResponse(url=response.url,body = response.text,encoding='utf-8')
    book = page.css('a::attr(title)').extract()
    return book
```

```python
book_scraper(url)
```

```python
def book_scraper_1(url):
    response = requests.get(url)
    page = TextResponse(url=response.url,body = response.text,encoding='utf-8')
    book = page.xpath('//a/@title').extract()
    return book 
```

```python
def price_scraper(url):
    response = requests.get(url)
    page = TextResponse(url=response.url,body = response.text,encoding='utf-8')
    price = page.css('p[class="price_color"]::text').extract()
    return price
```

```python
all_prices = []
for i in all_urls :
    all_prices.extend(price_scraper(i))
```
