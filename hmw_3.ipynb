```python
def most_frequent(date_1): 
    counter = 0
    num = date_1[0] 
      
    for i in date_1: 
        curr_frequency = date_1.count(i) 
        if(curr_frequency> counter): 
            counter = curr_frequency 
            num = i 
  
    return num 
  
print(most_frequent(date_1)) 
```
