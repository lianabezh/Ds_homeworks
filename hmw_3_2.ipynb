```python
def jobs_scraper(url):
    page = requests.get(url)
    response = TextResponse(url= page.url,body=page.text,encoding ="utf-8")
    jobs = response.xpath('//p[@class="font_bold"]/text()').extract()
    return jobs
```

```python
all_jobs  = []
i = 1 
while True :
    urls = "https://staff.am/en/jobs?page={}&per-page=50/".format(i)
    current_jobs = jobs_scraper(urls)
    if len(current_jobs) == 0:
        break
    else :
        all_jobs.extend(current_jobs)
        i +=1
        time.sleep(2)
            
print(all_jobs)
```

```python
def company_scraper(url):
    page = requests.get(url)
    response = TextResponse(url= page.url,body=page.text,encoding ="utf-8")
    companies  = response.xpath("//div[@class='job-inner job-item-title']/p[2]/text()").extract()
    return companies
```

```python
all_companies = []
i = 1 
while True :
    urls = "https://staff.am/en/jobs?page={}&per-page=50/".format(i)
    current_company = company_scraper(urls)
    if len(current_company) == 0:
        break
    else :
        all_companies.extend(current_company)
        i +=1
        time.sleep(2)   
        print(all_companies)
```

```python
highest_jobs_posters = Counter(all_companies)
```
